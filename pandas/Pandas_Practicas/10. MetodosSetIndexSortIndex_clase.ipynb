```python
import pandas as pd
```

```python
# leer y mostrar dataframe
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

# Crear Index

```python
import numpy as np
import random
```

```python
# crear valores no repetitivos para los index
new_index = np.arange(0,1000)
new_index
```

```python
# barajar/desordenar los indeces
random.shuffle(new_index)
new_index
```

```python
# crear una nueva columna con los nuevos index
df['new_index'] = new_index
df
```

# set_index()

```python
# fijar columna new_index como index
df.set_index('new_index', inplace = True)
df
```

# sort_index()

```python
# ordenar dataframe por new_index
df.sort_index()
```

```python
# orden descendiente por index
df.sort_index(ascending=False)
```
