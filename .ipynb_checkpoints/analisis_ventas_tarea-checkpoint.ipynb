# Análsis de ventas con NumPy

Supongamos que tienes datos de ventas de una tienda en diferentes meses y quieres analizarlos para obtener información clave, como el total de ventas, el promedio de ventas por mes y los meses con mejor desempeño.

## Paso 1. Importar NumPy y definir los datos


```python
import numpy as np

# Datos de ventas por cada mes
ventas = np.array([2300, 2500, 2600, 2100, 2900, 3100, 2800, 2700, 3000, 3300, 3500, 3700])
# Meses del año
meses = np.array(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
                  "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"])
print(ventas)
print(meses)
```

## Paso 2. Análisis de ventas con Numpy

Aplicaremos funciones clave para extraer información útil

### 1. Calcular el total y promedio de ventas al año.

```python
# Obtener el total de las ventas y el promedio, imprimirlas
total = np.sum(ventas)
print("Total de ventas en el año: $", total)

promedio = np.mean(ventas)
print("Promedio de ventas por mes: $", promedio)
```

### 2. Identificar el mes con más y menos ventas

```python
# Mes con ventas maximas
mes_maximo = meses[np.argmax(ventas)]
ventas_maximo = np.max(ventas)
print("Mes con mayores ventas: ", mes_maximo, " con $", ventas_maximo)
# Mes con ventas minimas
mes_minimo = meses[np.argmin(ventas)]
ventas_minimo = np.min(ventas)
print("Mes con menores ventas: ", mes_minimo, " con $", ventas_minimo)
```

### 3. Determinar los meses con ventas superiores al promedio

```python
# Meses con Ventas mayores al promedio
meses_superiores = meses[ventas > promedio]
ventas_superiores = ventas[ventas > promedio]
print(f"Promedio de ventas: {promedio}")
print("Meses con ventas superiores al promedio:")
for mes, venta in zip(meses_superiores, ventas_superiores):
    print(f"{mes}: ${venta} ")
```

## 4. Aplicar un incremento del 10% en las ventas

```python
ventas_incrementadas = ventas * 1.1
print("Ventas con incremento del 10%:", ventas_incrementadas)
```

## 5. Gráficar las ventas actuales y el incremento para el próximo año

```python
# importar la libreria para gráficar matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6)) # Especificando el tamaño

# Grafica de las ventas originales
plt.plot(meses, ventas, label="2024", marker='o', linestyle='-', color='green')

# Graficar las ventas con incremento
plt.plot(meses, ventas_incrementadas, label="2025", marker='o', linestyle='--', color='red')

# Agregar etiquetas

# Agregar las etiquetas
plt.xlabel("Meses")
plt.ylabel("Ventas ($)")
plt.title("Tendencia de ventas mensuales")

# Rotación de las etiquetas mes
plt.xticks(rotation=45)

# Agregar leyenda
plt.legend(loc = 'lower right')

# Mostrar el gráfico
plt.grid()
plt.show()
```

## 7. Detectar anomalias con un gráfico de caja (Boxplot)

```python
# Importamos la libreria seaborn
import seaborn as sns

# Graficar un boxplot para detectar anomalías
plt.figure(figsize=(10,6))
sns.boxplot(data=ventas)

# Agregar etiquetas
plt.title("Distribución de Ventas (Detección de Anomalías")
plt.ylabel("Ventas ($)")

# Mostrar el gráfico
plt.grid()
plt.show()

```
